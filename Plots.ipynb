```python
# Some plots for the paper
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# a bar graph with 7 bars
# values = [0.56,0.34, 0.29, 0.34,0.21 ,0.51, 0.48]
# bar names = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

def plot_bar(values, bar_names, title, xlabel, ylabel, filename):
    fig, ax = plt.subplots()
    ax.bar(bar_names, values)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    fig.savefig(filename)
    
    #make the exact numbers appear on the bars
    for i in range(len(values)):
        plt.text(i, values[i], str(values[i]), ha = 'center')
        
        
    plt.show()

plot_bar([0.56,0.34, 0.29, 0.34,0.21 ,0.51, 0.48], ['A', 'B', 'C', 'D', 'E', 'F', 'G'], 'Experiments test accuracy - Raw EEG dataser', 'Experiments', 'Test Accuracy', 'bar_graph.png')
```

```python
# Features dataset
values = [0.66,0.39, 0.35, 0.39,0.42 ,0.64, 0.57]
bar_names = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
plot_bar(values, bar_names, 'Experiments test accuracy - Features dataset', 'Experiments', 'Test Accuracy', 'bar_graph_features.png')
```
